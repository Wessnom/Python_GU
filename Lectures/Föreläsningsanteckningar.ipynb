```python
s = 'sträng'

print(s[1:5])   # This will print the characters from index 1 to 4
print(s[1:5:2]) # This will print every second character from index 1 to 4
print(s[:5])    # This will print from the start to index 4
print(s[4:])    # This will print from index 4 to the end
print(s[-6:-1]) # This will print from index -6 to -1
print(s[-1::-1])# This will print the string in reverse order
print(s[8:10])  # This will print an empty string since the slice is out of range

print(ord('A'))  # This will print the Unicode code point of 'a'
```
