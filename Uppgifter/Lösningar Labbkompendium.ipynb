##### Andragradsekvation i Python

Skapa ett program som hittar och skriver ut lösningen till  
andragradsekvationen

\[
ax² + bx + c = 0
\]

där **a**, **b** och **c** är koefficienter som anges av användaren.  
Vi utgår från att ekvationen har **två reella lösningar**.

```python
import math

a = float(input("Ange a: "))
b = float(input("Ange b: "))
c = float(input("Ange c: "))

p = b / a
q = c / a

x1 = -p/2 + math.sqrt((p/2)**2 - q)
x2 = -p/2 - math.sqrt((p/2)**2 - q)

print(x1, x2)
```

Skapa två heltal av datatypen int. Skriv ett program som skriver ut  


a) Dessa tal formaterade som hexadecimala tal. 


b) Kvoten  av  de  två  talen.  Resultatet  ska  skrivas  ut  som  ett  flyttal  med  3  siffror  efter 
decimalpunkten. 


c) Resten efter division mellan de två talen.

```python
# Initierar två heltal (int)
a = 10
b = 15

# a) Skriv ut talen som hexadecimala
print(f"a i hexadecimalt: {hex(a)}")
print(f"b i hexadecimalt: {hex(b)}")

# b) Skriv ut kvoten som flyttal med 3 decimaler
kvot = a / b
print(f"Kvoten a/b: {kvot:.3f}")

# c) Resten efter division
rest = a % b
print(f"Resten efter division: {rest}")
```

Uppgift 1.1.3 (1p):

Skriv ett program som läser in information om människor, så som namn, 
efternamn, och födelseår, angett av användaren och skriver ut en formaterad tabell med denna 
information. Man får inte använda tabulate modulen.

```python
# Lista för att lagra personer
people = []

antal_personer = int(input("Ange antal personer att registrera: "))

for _ in range(antal_personer):
    namn = input("Ange förnamn: ")
    efternamn = input("Ange efternamn: ")
    dob = input("Ange födelseår: ")
    people.append((namn, efternamn, dob))
    
print("\nFörnamn; Efternamn; Födelseår")
for p in people:
        print(f"{p[0]}; {p[1]}; {p[2]}")
```
