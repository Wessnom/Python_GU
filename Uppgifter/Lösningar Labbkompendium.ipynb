##### Uppgift 1.1.1 (1p): Skapa ett program som hittar och skriver ut lösningen till 
andragradsekvationen ax2 + bx + c = 0.
a, b, och c är koefficienter som anges av användaren. 
Vi utgår från att ekvationen har två reella lösningar.

```python
import math

# Läs in koefficienterna från användaren
a = float(input("Ange koefficienten a: "))
b = float(input("Ange koefficienten b: "))
c = float(input("Ange koefficienten c: "))

# Beräkna diskriminanten
diskriminant = b**2 - 4*a*c

# Beräkna lösningarna med andragradsekvationens formel
x1 = (-b + math.sqrt(diskriminant)) / (2*a)
x2 = (-b - math.sqrt(diskriminant)) / (2*a)

# Skriv ut resultatet
print(f"\nLösningarna till ekvationen {a}x² + {b}x + {c} = 0 är:")
print(f"x₁ = {x1}")
print(f"x₂ = {x2}")
```
