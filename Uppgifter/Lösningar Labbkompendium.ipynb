##### Uppgift 1.1.1 (1p): Skapa ett program som hittar och skriver ut lösningen till 
andragradsekvationen ax2 + bx + c = 0.
a, b, och c är koefficienter som anges av användaren. 
Vi utgår från att ekvationen har två reella lösningar.

```python
import math

a = float(input("Ange a: "))
b = float(input("Ange b: "))
c = float(input("Ange c: "))

p = b / a
q = c / a

x1 = -p/2 + math.sqrt((p/2)**2 - q)
x2 = -p/2 - math.sqrt((p/2)**2 - q)

print(x1, x2)
```
